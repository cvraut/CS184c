Find shortest path from West Coast to East Coast (Start -> Finish) using DP or viterbi

```python
towns_for_day = {"Start"}
DP_dict = defaultdict(lambda : (inf,"Prev"))
DP_dict["Start"] = (0,"SRC") # this holds shortest dist to town at "key" from west coast & previous town in traceback
while len(towns_for_day) > 0:
    new_towns = set()
    for town in towns_for_day:
        cur_dist = DP_dict[town][0]
        dests = mygraph.adj[town]
        for dest,obj in dests.items():
            new_towns.add(dest)
            route_dist = obj['weight']
            if cur_dist+route_dist < DP_dict[dest][0]:
                DP_dict[dest] = (cur_dist+route_dist,town)
        
    towns_for_day = new_towns
path = []
cur = "End"
while cur != "SRC":
    path.append((DP_dict[cur][0],cur))
    cur = DP_dict[cur][1]

print("Total Distance = {}\nPath:".format(path[0][0]))
for i in range(-1,-1*len(path)-1,-1):
    day_i = -1*i-2
    if day_i < 0 or day_i >= len(path)-2:
        day_i = '-'
    dist,loc = path[i]
    print("{:15s} - Day {}, total distance traveled {:4d} miles".format(loc,day_i,dist))
```
